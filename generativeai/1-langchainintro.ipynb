### Streaming and batch processing
in simple words stream means data comes in chunks not the whole data loaded at the first time in langchain and generateive ai application it something like that when the llm model generate something then line by line it print the output not wait for the whole output generate as the things are generated it will show the output it make the user interface  perfect  it show the data in real time chunks

```python
response = model.invoke("Write me an eassay with 1000 words")
response.content
#almost take 6 second first generate all the data and then it show
```

```python
chuks = model.stream("Write me a  1500 words paragraph ai")
for mytes in chuks:
    print(mytes.text, end="", flush=True)
```

```python
for chunks in model.stream("Why parratos have different features"):
    print(chunks.text, end="", flush=True)
```

###     Batch
batch is a collection of independent request that run parellel or we can say that multiple input process together

```python
responses = model.batch([
    "why do parratos have colorgul features",
    "how do the airoplane fly",
    "what is the main motive of a person"
],
config={
    'max_concurrency':5,
})

for response in responses:
    print(response)
```
