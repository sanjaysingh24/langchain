### Tools
why we need tools (so in llm there are a training cut on the models in simple language we can say that the model are trained a  specific date or year than for example if the model was trained till the oct 2025 and i ask the input to the llm for the latest ai news in nov and dec 2025 than at that time llm are not trained over the data nov 2025 and dec 2025 but llm  create a output or  hallustion which is not that much accurate) so here we introduce the tools when i give the input of the current news in ai of dec 2025 then what it will do first it goes to the llm and check which tool i should use for this if the tool is present  then call this particular tool and  take some context from the tool and when llm recieve the context then it will add the prompt to the context and generate the output to the user 

models can request to call a tools that perform task such as  fetching data from database ,searching web, and running code tools are pairing of:
1- a schema, including the name  of the tool, a description, and/or argument definition(often  a Json Schema)
2- a function or  coroutine to excute

```python
import os
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"] = os.getenv("GROK_API_KEY")
model = init_chat_model("groq:llama-3.1-8b-instant")
response = model.stream("write me an essay of ai")
for r in response:
    print(r.text,end="",flush=True)
```

```python
#tools
from langchain.tools import tool

@tool
def get_current_weather(location:str)->str:
    """ get the weather at a location"""
    return f"it's sunny in {location}"
modles_with_tools = model.bind_tools([get_current_weather])
```

```python
response = modles_with_tools.invoke("What is the weather in banglore")
response.tool_calls
```

### Tool Execution Loop

```python
#step 1: model generates tool calls
message = [{"role":"user","content":"what is the weather in banglore?"}]
ai_msg= modles_with_tools.invoke(message)


message.append(ai_msg)

#step 2: Execute tools  and collect results
for tool_call in ai_msg.tool_calls:
    #execute the tool with the generated arguments
    tool_result = get_current_weather.invoke(tool_call)
    
    message.append(tool_result)


final_response = modles_with_tools.invoke(message)
print(final_response.text)
```
